```python
# USe of isin() function - > this function is used to check a list element from the serise data
#Q  find the most Destructive palyer in death overs  of ipl
  # finding the strik rate = (run/balls)*100;
    # batsmans should have played min 200 balls in 16 -20 th over of IPL
    
    
# firstly we filter the dataset to find the batsman played in 16 - 20 th overs
Deathover=Data1["over"]>15
overs=Data1[Deathover]       #  - it return the dataframe that contains only 16 to 20 th over matches

# secondly we need to find those batsman who have played at least 200 balls in 16 -20 th overs

ball_played =overs.groupby("batsman")["total_runs"].count() 
mask=ball_played>200
ball_200=ball_played[mask].sort_values(ascending=False)
Batsman_list=ball_200.index.tolist()
Batsman_list
```

```python
final=overs[overs["batsman"].isin(Batsman_list)]
# balls
ball=final.groupby("batsman")["batsman_runs"].count()
# Runs
runs=final.groupby("batsman")["batsman_runs"].sum()
((runs/ball)*100).sort_values(ascending=False).head(10).plot(kind="pie")
```

```python
# *IMPORTANT* Q find the orange cap holder for every season;;
Data1
Data2

# NOTE: since we need to have a single data frame that contain seasons and runs made by batsmans , as we dont have the same .
# So , we will merge the both table ( condition atleat one coloum should be common)

new=Data1.merge(Data2,left_on="match_id",right_on="id")
new  # A new data frame that contain both the season and runs coloum ...
#new.shape

#we need to group by 
new.groupby(["season","batsman"])["total_runs"].sum()  #NOTE :: It return Multiple index serise 
```

```python
new.groupby(["season","batsman"])["total_runs"].sum().sort_values(ascending=False).reset_index().drop_duplicates("season",keep="first").sort_values("season")[["season","batsman"]]
# reset_index -> it convert the sersie data into dataFrame type 
# we used drop_duplicates to remove all the other rows other than top RUNS scrocer of that seasons. 
```

```python
# THE MOST IMPORTANT TOOL-> Pivote Table 

# A pivoet table is ana intracrive way to quickly summarize large amount of data . You can use a pivote Table to analyze numeical data in detail , and can answer unanticipated 
# Question about your data ..

# Signature: Data1.pivot(*, index=None, columns=None, values=None) -> 'DataFrame'
# Docstring:
# Return reshaped DataFrame organized by given index / column values.


# index-> it is the index of pivote table ( MULITPILE Can ALSO EXIST )
# coloum-> IT Denotes the coloum of pivote table 
# values-> index ki coloum ki value kya hogi ? 
# aggregate-> value kya hoga mean , sum etc

# using it over the IPL data set  -> we need to find that every team in 1-20 th overs me kitne 6's kiye hai ..

mask=Data1['batsman_runs']==6
new=Data1[mask]

pt=new.pivot_table(index="over",columns="batting_team",values="batsman_runs",aggfunc="count")
```
